```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
from sklearn.metrics import pairwise_distances
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
%matplotlib inline
```

```python
df = pd.read_csv('../CSVs/hindu_articles.csv')
print(df.shape)
df.head(10)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tf_idf = vectorizer.fit_transform(df['body'].values.astype('U'))
```

```python
tf_idf
```

```python
model = NearestNeighbors(metric='euclidean', algorithm='brute', n_neighbors=20)
model.fit(tf_idf)
```

```python
distances, indices = model.kneighbors(tf_idf[4])
```

```python
neighbors = pd.DataFrame({'distances':distances.flatten(), 'indices':indices.flatten()})
neighbors.join(df, on='indices')
```
